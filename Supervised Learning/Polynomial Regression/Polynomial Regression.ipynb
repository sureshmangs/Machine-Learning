```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
dataset.head()
```

```python
X = dataset.iloc[:, 1:2].values
Y = dataset.iloc[:, -1].values
```

```python
# Fitting Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, Y)
```

```python
# Fitting Polynomial Regression to the Training set
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, Y)
```

```python
# Visualizing the Linear Regression results
plt.scatter(X, Y, color = 'red', label = 'Scatter Plot')
plt.plot(X, lin_reg.predict(X), color = 'blue', label = 'Line Plot')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Visualizing the Polynomial Regression results
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X, Y, color = 'red', label = 'Scatter Plot')
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), color = 'blue', label = 'Line Plot')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Predicting a new result Linear Regression
lin_reg.predict(np.array(6.5).reshape(1,-1))
```

```python
# Predicting a new result polynomial Regression
lin_reg2.predict(poly_reg.fit_transform(np.array(6.5).reshape(1,-1)))
```
